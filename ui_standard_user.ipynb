```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options  
from selenium.webdriver.support.ui import Select

```

```python
options = Options()
options.add_argument("--start-maximized")  
service = Service("chromedriver.exe") 
driver = webdriver.Chrome(service=service, options=options)

try:
    driver.get("https://www.saucedemo.com")
    print("Página acessada com sucesso!")
except Exception as e:
    print(f"Erro ao acessar a página: {e}")
    driver.quit()
    exit()

try:
    # Localiza os campos de usuário e senha
    username_input = driver.find_element(By.ID, "user-name")
    password_input = driver.find_element(By.ID, "password")
    print("Campos de login localizados!")
except Exception as e:
    print(f"Erro ao localizar os campos de login: {e}")
    driver.quit()
    exit()

try:
    username_input.send_keys("standard_user") # Usuário Padrão
    password_input.send_keys("secret_sauce")
    print("Credenciais preenchidas!")
    password_input.send_keys(Keys.RETURN) 
    print("Formulário de login enviado com sucesso!")
except Exception as e:
    print(f"Erro ao inserir dados de login: {e}")
    driver.quit()
    exit()

try:
    driver.implicitly_wait(10)
    print("Login realizado com sucesso!")
except Exception as e:
    print(f"Erro ao realizar login: {e}")
    driver.quit()
    exit()

try:
    # Adiciona itens ao carrinho
    btn_add_to_cart_1 = driver.find_element(By.ID, "add-to-cart-sauce-labs-backpack")
    btn_add_to_cart_1.click()
    print("Item 1 adicionado ao carrinho!")

    btn_add_to_cart_2 = driver.find_element(By.ID, "add-to-cart-sauce-labs-fleece-jacket")
    btn_add_to_cart_2.click()
    print("Item 2 adicionado ao carrinho!")

    btn_add_to_cart_3 = driver.find_element(By.ID, "add-to-cart-test.allthethings()-t-shirt-(red)")
    btn_add_to_cart_3.click()
    print("Item 3 adicionado ao carrinho!")
except Exception as e:
    print(f"Erro ao adicionar itens ao carrinho: {e}")
    driver.quit()
    exit()

try:
    # Remover itens ao carrinho
    btn_remove_to_cart = driver.find_element(By.ID, "remove-sauce-labs-fleece-jacket")
    btn_remove_to_cart.click()
    print("Item 2 removido do carrinho!")
except Exception as e:
    print(f"Erro ao adicionar itens ao carrinho: {e}")
    driver.quit()
    exit()

try:
    filter_dropdown = driver.find_element(By.CSS_SELECTOR, '[data-test="product-sort-container"]')
    select = Select(filter_dropdown)
    select.select_by_value("za")
    print("Filtro (Z to A) aplicado com sucesso!")
except Exception as e:
    print(f"Erro ao aplicar o filtro (Z to A): {e}")

try:
    filter_dropdown = driver.find_element(By.CSS_SELECTOR, '[data-test="product-sort-container"]')
    select = Select(filter_dropdown)
    select.select_by_value("hilo")
    print("Filtro (high to low) aplicado com sucesso!")
except Exception as e:
    print(f"Erro ao aplicar o filtro (high to low): {e}")

try:
    filter_dropdown = driver.find_element(By.CSS_SELECTOR, '[data-test="product-sort-container"]')
    select = Select(filter_dropdown)
    select.select_by_value("az")
    print("Filtro (A to Z) aplicado com sucesso!")
except Exception as e:
    print(f"Erro ao aplicar o filtro (A to Z): {e}")

try:
    filter_dropdown = driver.find_element(By.CSS_SELECTOR, '[data-test="product-sort-container"]')
    select = Select(filter_dropdown)
    select.select_by_value("lohi")
    print("Filtro (low to high) aplicado com sucesso!")
except Exception as e:
    print(f"Erro ao aplicar o filtro (low to high): {e}")

try:
    carrinho = driver.find_element(By.CSS_SELECTOR, '[data-test="shopping-cart-link"]')
    carrinho.click()
    print("Carrinho acessado com sucesso!")
except Exception as e:
    print(f"Erro ao acessar carrinho: {e}")

try:
    btn_checkout = driver.find_element(By.ID, "checkout")
    btn_checkout.click()
    print("Redirecionado para Checkout: Your Information!")
except Exception as e:
    print(f"Erro ao redirecionar para Checkout: Your Information: {e}")
    driver.quit()
    exit()

try:
    # Localiza os campos
    firstname_input = driver.find_element(By.ID, "first-name")
    lastname_input = driver.find_element(By.ID, "last-name")
    postalcode_input = driver.find_element(By.ID, "postal-code")
    print("Campos de checkout localizados!")
except Exception as e:
    print(f"Erro ao localizar os campos de checkout: {e}")
    driver.quit()
    exit()

try:
    # Preenche os campos com as credenciais
    firstname_input.send_keys("Iara")
    lastname_input.send_keys("Silva")
    postalcode_input.send_keys("36301-180")
    print("Credenciais preenchidas!")
    btn_continue = driver.find_element(By.ID, "continue")
    btn_continue.send_keys(Keys.RETURN)
    print("Formulário enviado com sucesso!")
except Exception as e:
    print(f"Erro ao inserir dados: {e}")
    driver.quit()
    exit()

try:
    btn_finish = driver.find_element(By.ID, "finish")
    btn_finish.click()
    print("Compra finalizada com sucesso!")
    print("Teste Passou!")
except Exception as e:
    print(f"Erro ao finalizar compra: {e}")
    driver.quit()
    exit()

# Finaliza o driver após o sucesso
driver.quit()
```
